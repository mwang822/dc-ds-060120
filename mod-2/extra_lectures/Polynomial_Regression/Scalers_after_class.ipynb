```python
import numpy as np

```

```python
np.random.seed(seed=18032020)

## cov: Covariance matrix: How columns are related?
X = np.random.multivariate_normal(mean= (0.001,1),cov = np.array([[0.1, 0.70], [0.70, 10]]), size = 200)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[:,0], X[:,1])
plt.xlim(-5,10)
plt.xlabel('X1 variable')
plt.ylabel('X2 variable')
plt.show()
```

## Initial Distribitions of the Columns


```python
plt.figure(figsize= (10,6))
plt.subplot(1,2,1)
plt.title('X1')
plt.hist(X[:,0])

plt.subplot(1,2,2)
plt.title('X2')
plt.hist(X[:,1])

plt.show()
```

## Scalers in Sklearn and their effect on the data

## Centralize

```python
X_centralized = X - X.mean()
plt.xlim(-7.5, 10)
plt.scatter(X_centralized[:,0], X_centralized[:,1])

plt.title('Centralizing the data')

plt.xlabel('X1_scaled: Mean = 0')
plt.ylabel('X2_scales: Mean = 0')

plt.show()
```

```python
X_centralized[:,1].std()
```

## StandardScaler


!! Sensitive to the outliers!!

[Sklearn Documentation](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler)

```python
from sklearn.preprocessing import StandardScaler
```

```python
## instantiate
standard_scaler = StandardScaler()

## fit and transform

## fit: Learn mean and std deviation of the data
## transform apply z-score formula to the data
X_std_scaled = standard_scaler.fit_transform(X)

plt.scatter(X_std_scaled[:,0], X_std_scaled[:,1])

plt.title('Scaling with Standard Scaler')

plt.xlabel('X1_scaled: Mean = 0, std = 1')
plt.ylabel('X2_scales: Mean = 0, std = 1')

plt.show()
```

```python
plt.figure(figsize= (10,6))
plt.subplot(1,2,1)
plt.title('X1')
plt.hist(X_std_scaled[:,0])

plt.subplot(1,2,2)
plt.title('X2')
plt.hist(X_std_scaled[:,1])

plt.show()
```

```python
## let's investigate columns separately

X1_std_scaled_mean = round(X_std_scaled[:,0].mean(), 4)

X1_std_scaled_std = round(X_std_scaled[:, 0].std(), 4)

print('Mean of X1 after standard scaling is %.4f'%X1_std_scaled_mean)
print('Std of X1 after standard scaling is %.4f'%X1_std_scaled_std)
```

```python
## let's investigate columns separately

X2_std_scaled_mean = round(X_std_scaled[:,1].mean(), 4)

X2_std_scaled_std = round(X_std_scaled[:, 1].std(), 4)

print('Mean of X2 after standard scaling is %.4f'%X2_std_scaled_mean)
print('Std of X2 after standard scaling is %.4f'%X2_std_scaled_std)
```

## Normalizer

!!! Not a scaler!!!!

[Sklearn Documentation](https://scikit-learn.org/stable/modules/preprocessing.html#normalization)


- Normalizer scaler is useful when we apply TF-idf algorithm in Natural Language Processing

- More generally it is useful when we have sparse dataset (Not dense)

```python
from sklearn.preprocessing import Normalizer
```

```python
normalizer = Normalizer()

X_normalized = normalizer.fit_transform(X.T ).T
```

```python
plt.scatter(X_normalized[:,0], X_normalized[:,1])

plt.title('Scaling with Normalizer')

plt.xlabel('X1_normalized: Mean = 0, std = 1')
plt.ylabel('X2_norm: Mean = 0, std = 1')

plt.show()
```

```python
X_normalized[:,0].mean()
X_normalized[:,0].std()
```
